```python
import pandas as pd 
import os
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

import numpy as np 



year_and_quarter = [] 
measurement = []
```

```python

for file in os.listdir("Time Series Analysis Data"): 
    path = f"Time Series Analysis Data/{file}"
    print(path)
    df = pd.read_csv(path)
    df = df.dropna()
    df = df[df["Product Name"].isin(["LUPRON DEPOT", "LUPRON DEP"])]
    results = df.groupby("Quarter", as_index = False)["Medicaid Amount Reimbursed"].sum()
    year = file.split("-")[1]
    print(results)
    for qtr in results["Quarter"]: 
        label = year +"Q" + str(qtr)
        year_and_quarter.append(label)
    for item in results["Medicaid Amount Reimbursed"]: 
        measurement.append(item)
   
        
        
    




```

```python
import matplotlib.pyplot as plt 


plt.plot(year_and_quarter, measurement)
```

```python
# ACF AND PACF PLOTS 

measurement = np.array(measurement)
```

```python
plot_pacf(measurement, title = "PACF plot")
```

```python
plot_acf(measurement, title = "ACF plot")
```

These plots suggest that an AR(2) plot might be best for predicting Medicaid's quaterly spending on Lupron Depot. 

```python
training_data =measurement[:int(.8 * len(measurement))]
test_data = measurement[int(int(.8 * len(measurement))):]
```

```python
labels_training_data = year_and_quarter[:int(.8 * len(year_and_quarter))]
labels_test_data = year_and_quarter[int(int(.8 * len(year_and_quarter))):]
```

```python
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.api import AutoReg 


ar_model = AutoReg(measurement, lags = 2, trend = 'n')


ar_results = ar_model.fit()
```

```python
ar_results.summary()
```

The roots appear to be on or above the unit circle so we are going to consider this a stable model. 

```python
ar_results.plot_diagnostics()
```

The model seems to capture the trends of the data well according to the Correlogram (Maybe Include some seasonality?)

```python
pred = ar_results.predict(start=len(training_data), end=len(training_data) + len(test_data) - 1)
pred = pd.DataFrame({"Date": labels_test_data, "pred":pred})
pred.set_index("Date", inplace = True)
plt.plot(pred, label = "Predictions")
plt.plot(test_data, label = "Actual")
plt.legend()
plt.title("Actual vs Predictions Plot")
plt.tick_params(axis = "x", labelrotation = -35)
```

X Ticks Formatting: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tick_params.html#matplotlib.pyplot.tick_params

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

print("MAE", mean_absolute_error(pred, test_data))
print("RMSE", np.sqrt(mean_squared_error(pred,test_data)))
```

The values are quite large, assumingly because the data deals with large quantities of numbers, maybe add some seasonality?
